## $\text{Marco Te\'{o}rico}$

#### $\rightarrow \text{Media} \leftarrow$

Representa el valor promedio de los datos y se calcula sumando todos los valores y dividiéndolos por el número total de observaciones. La media es útil para obtener una visión general de la magnitud típica de los valores en un conjunto de datos. (Economipedia., s.f.)
$$
\mu=\sum{\frac{x_i}{n}\,_{(1)}}
$$

#### $\rightarrow \text{Desviaci\'{o}n Est\'{a}ndar}\leftarrow$

Mide la variabilidad o dispersión de los datos alrededor de la media. Es crucial para evaluar la variabilidad de los datos. (SAS Institute Inc., s.f.)
$$
\sigma=\sqrt{\frac{\sum(x_i-\mu)^2}{n}}\,_{(2)}
$$

#### $\rightarrow \text{SEM (Error Est\'{a}ndar de la Media)} \leftarrow$

El Error Estándar de la Media (SEM) es una medida que describe cuán precisamente la media de una muestra representa la media de la población general. 
El SEM se calcula dividiendo la desviación estándar de la muestra por la raíz cuadrada del número de observaciones en la muestra. (INTEF., s.f.)
$$
SEM = \frac{\sigma}{\sqrt{n}}\, _{(3)}
$$

#### $\rightarrow \text{SEM \% of Mean (Error Est\'{a}ndar de la Media como Porcentaje de la Media)} \leftarrow$

El SEM % of Mean también expresa el SEM como un porcentaje de la media.Este valor proporciona una medida de cuán significativa es la incertidumbre alrededor de la media en términos relativos al tamaño de la media misma. Es útil para comparar la precisión de las medias entre diferentes conjuntos de datos donde las escalas de medición pueden diferir. (INTEF., s.f.)

$$
\text{SEM \% of Mean} = \frac{SEM}{\mu} \times 100\, _{(4)}
$$

#### $\rightarrow \text{Rango Intercuartilico}\leftarrow$

El IQR es una medida de dispersión estadística que se utiliza para describir la variabilidad de los datos al concentrarse en la parte central de una distribución. El IQR es la diferencia entre el tercer cuartil (Q3) y el primer cuartil (Q1) de los datos. Estos cuartiles dividen el conjunto de datos en cuatro partes iguales. (Economipedia., s.f.)
$$
\text{IQR} = Q3 - Q1\,_{(5)}
$$

## $\text{Resultados y Discusi\'{o}n de Resultados}$

### $\rightarrow \text{ Resultados } \leftarrow$

Primero se importaron las bibliotecas y se cargó la data proveniente de un archivo .csv al objeto tipo DataFrame de pandas data_chf.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from scipy.stats import iqr


data_chf = pd.read_csv('Data_CHF_Zhao_2020_ATE.csv')
data_chf
```

Seguidamente se aplicó el método describe para que se proporcionaran medidas estadísticas a priori de la data:

```python
data_chf.describe()
```

Luego se creó una lista que solo incluye las variables numéricas del DataFrame.

```python
numerical_cols = []

numerical_cols = data_chf[
    [
    'pressure [MPa]', 'mass_flux [kg/m2-s]', 'x_e_out [-]',
    'D_e [mm]', 'D_h [mm]', 'length [mm]', 'chf_exp [MW/m2]'
]
    ]
numerical_cols
```

Seguidamente se calcularon las medidas de dispersión anteriormente mencionadas para cada una de las variaables numéricas, guardándolas en una lista, respectivamente.

```python
# Desviación Estándar

std_devs = list(numerical_cols.std())
std_devs
```

```python
# Media

means = list(numerical_cols.mean())
means
```

```python
# Cantidad de Muestras para cada Variable

n_samples = len(numerical_cols)
n_samples
```

```python
# Error Estándar Medio (SEM)

sems = list((numerical_cols.std()/np.sqrt(len(numerical_cols))))
sems
```

```python
# Porcentaje del Error Estándar respecto de la Media

sem_percentage = list(((numerical_cols.std()/np.sqrt(len(numerical_cols))))*(100/numerical_cols.mean()))
sem_percentage
```

```python
# Rango Intercuartil

iqr_data = list(iqr(numerical_cols, axis=0))
iqr_data
```

Luego se hizo un diccionario con todas las anteriores medidas para crear un DataFrame a partir de este.

```python
data = {
    'Media' : means,
    'Desviación Estándar' : std_devs,
    'ESM' : sems,
    'ESM % de la Media' : sem_percentage,
    'Rango Cuartílico' : iqr_data
}

data
```

Se creó el DataFrame con el diccionario, cambiando los índices por los nombres de las columnas del DF numerical_cols.

```python
df = pd.DataFrame(data)
df.set_index(numerical_cols.columns, inplace=True)
df
```

Se hizo la gráfica de todas las medidas obtenidas a través de un ciclo for.

```python
#Se crea el fondo de blanco de la figura con Seaborn.
sns.set(style="whitegrid")

# Se crea la figura y los subgráficos. Creamos una matriz de 3x3 gráficos y asignamos su tamaño.
fig, axes = plt.subplots(3, 3, figsize=(18, 12))

# Se aplana la matriz de los ejes para facilitar la iteración
axes = axes.flatten()

for i, col in enumerate(numerical_cols):  # Enumerar las columnas numéricas
    sns.histplot(numerical_cols[col], kde=True, ax=axes[i])  # Graficar en el eje correspondiente
    axes[i].set_title(f'Histograma de {col}')  # Agregar título al gráfico

# Ajustar el diseño para evitar superposición de gráficos
plt.tight_layout()

# Mostrar los gráficos
plt.show()
```
